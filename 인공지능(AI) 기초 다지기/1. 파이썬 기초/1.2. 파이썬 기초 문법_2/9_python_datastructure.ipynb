## 스택 (stack)
- 나중에 넣은 데이터를 먼저 반환하도록 설계된 메모리 구조
- Last In First Out (LIFO)
- Data의 입력을 push, 출력을 pop이라고 함.


### stack with list object
- 리스트를 사용하여 스택 구조를 구현 가능
- push를 append(), pop을 pop()를 사용

```python
a = [1,2,3,4,5]
a.append(10)
a.append(20)
a.pop() # 20 출력
```

```python
a.pop() # 10 출력
```

### stack example

```python
word = input("Input a word: ") # input word
word_list = list(word)         # string to list 형변환
for i in range(len(word_list)):
    print(word_list.pop())     # 하나씩 빼면서 출력
```

#### 일반적으로 너가 c언어로 배운 stack 자료구조, FILO에 대해선
#### push라는 함수와 pop이라는 함수를 별도로 구현해서 사용하거나
#### 아니면 main 안에 직접 풀어서 사용하는 형식으로 썼으나,
#### python은 아주 편리하게 이미 push, pop함수가 구현되서 라이브러리에 들어가 있기에 이렇게 사용할 수 가 있는 것이다.

```python
a = [1,2,3,4,5]
a.append(10)
a.append(20)
a.pop(0) # 1 출력
```

## deque
- stack과 queue를 지원하는 모듈
- list에 비해 효율적인 = 빠른 자료 저장 방식을 지원함
- queue의 반대 개념....?

```python
from collections import deque

deque_list = deque()
for i in range(5):
    deque_list.append(i)
print(deque_list)
```

```python
deque_list.appendleft(10)
print(deque_list)
```

- rotate, reverse 등 linked list의 특성을 지원함.
- 기존 list 형태의 함수를 모두 지원함.
- 너 옛날에 자료구조할 때 트리 구조로 배울 때 queue, deque다 배웠자나....?


- 기존 list 형태의 함수를 모두 지원함

```python
deque_list.rotate(2)
print(deque_list)
```

```python
deque_list.rotate(2)
print(deque_list) ## 뒤의 두 칸 rotate 시킨다.
```

```python
print(deque(reversed(deque_list))) ## reversed시킨다는 함수의 사용
```

```python
deque_list.extend([5,6,7])
print(deque_list)
```

```python
deque_list.extendleft([5,6,7])
print(deque_list)
```

- deque는 기존 list보다 효율적인 자료구조를 제공
- 효율적 메모리 구조로 처리 속도 향상

## defaultdict
- dict type의 값에 기본 값을 지정, 신규값 생성시 사용하는 방법


- dict type의 값에 기본 값을 지정, 신규값 생성시 사용하는 방법

```python
from collections import defaultdict
d = defaultdict(object) # default dictionary를 생성
d = defaultdict(lambda: 0) # default 값을 0으로 설정함.
print(d["first"])
```

- 하나의 지문에 각 단어들이 몇 개나 있는지 세고 싶을 경우?
- 이 때 쓸 수 있는게 defaultdict이다.
- Text-mining 접근접 - Vector Space Model

```python
text = """A press release is the quickest and easiest way yo get free publicity. if well written, a press release can result in multiple published articles about your firm and its porducts. And that can mean new prospects contacting you asking you to sell to them. ***.""".lower().split()
print(text)
```

```python
from collections import OrderedDict
word_count = defaultdict(lambda: 0) # default값을 0으로 설정함.
for word in text:
    word_count[word] += 1
for i, v in OrderedDict(sorted(word_count.items(), key=lambda t: t[1], reverse=True)).items():
    print(i,v)
```

## counter
- Sequence type의 data element들의 갯수를 dict 형태로 반환

```python
from collections import Counter

c = Counter() # a new, empty counter
c = Counter('gallahad') # a new counter from an iterable
print(c)
```

- dict type, keyword parameter 등도 모두 처리 가능

```python
c = Counter({'red':4, 'blue':2}) # a new counter from a mapping
print(c)
```

```python
print(list(c.elements()))
```

```python
c = Counter(cats=4, dogs=8) # a new counter from keyword args
print(c)
print(list(c.elements()))
```

- set의 연산들을 지원함

```python
c = Counter(a=4, b=2, c=0, d=-2)
d = Counter(a=1, b=2, c=3, d=4)
c.subtract(d) # c - d
print(c)
```

```python
print(c + d)
print(c & d)
print(c | d)
```
