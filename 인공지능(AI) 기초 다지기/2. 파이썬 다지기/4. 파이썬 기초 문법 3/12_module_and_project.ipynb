## Module 만들기
- 파이썬의 Module == py 파일을 의미
- 같은 폴더에 Module에 해당하는 .py파일과 사용하는(실행시키는) .py를 저장한 후
- import 문을 사용해서 module을 호출

- 기본적으로 가장 중요한 것은 둘 다 같은 폴더 안에 있어야 한다!

## namespace


- 그냥 import로 그 모듈 파이썬 파일을 불러오면 안에 있는 내용을 싹 다 가져오니까
- from-import를 통해 쓸 것만 가져온다는 것.

## 패키지

- init파일은 각 폴더 별로 반드시 기초필수적으로 필요한 놈이다.

- from ~ import ~ 에서 from 뒤에는 파일(모듈)이 올 수도, root-tree관계가 되있는 폴더를 가져올 수 도 있다. 반드시 사전"설계"가 되어 있어야 한다.


- 즉 3.3+부터는 안해도 되는 것.
- 그러나 init은 반드시 필요하다는게 현업에서의 룰
- 그러니 반드시 만드는 걸로 인식하고 있는 것이 좋다.

- 이 패키지를 공유하면 그게 프로젝트이다.
